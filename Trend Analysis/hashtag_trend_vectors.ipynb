```python
reverse_tag_groups = {tag: cluster for cluster, tags in tag_groups.items() for tag in tags}


trend_data = trend_data[trend_data['hashtag'].isin(reverse_tag_groups.keys())]

# Create a new column 'cluster' in the DataFrame using the reverse_tag_groups dictionary
trend_data['cluster'] = trend_data['hashtag'].map(reverse_tag_groups)

# Convert the 'date' column to datetime format
trend_data['date'] = pd.to_datetime(trend_data['date'])

# Group by 'cluster' and 'date' and count the occurrences
grouped_df = trend_data.groupby(['cluster', 'date']).size().reset_index(name='count')

# Display the result
grouped_df.head()
```
